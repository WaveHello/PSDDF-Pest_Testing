```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

The purpose of this code is to plot equations used in PSDDF, if I want more clarity.

```python
h = 10 # arb. height
C_alpha = 2 # Secondary compression index
e0 = 5
tp = 1000 # 

t = np.linspace(tp, 50*tp)
Ss = C_alpha/(1+e0) * h * np.log(t/tp)

plt.plot(t, Ss)
plt.show()
```
