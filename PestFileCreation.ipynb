When selecting columns in Pest indexing starts at 1 and is inclusive of the last index. For example for a string "1236.567" that starts on the far left of a file, the selection would be [1:8]. For python the syntax would be [0:8] since python is not inclusive of the second index.

```python
test=  "     1700.69629       -0.96377"
print(len(test))
print(test[22:30])

test2 = "     1700.69629       71.26960"
print(test2[22:])

test3 = "    1.00            1.21072"
test3[20:27]

test4 = "1236.5678495.000-900.000"

test4[0:8]
```
